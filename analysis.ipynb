```python
import pandas as pd

# Load CSV file
df = pd.read_csv("sales.csv")

# Display first few rows
print(df.head())
```

```python
sales_summary = df.groupby('Product')['Sales'].sum().reset_index()
print(sales_summary)
```

```python
import matplotlib.pyplot as plt

# Bar chart of sales by product
sales_summary.plot(
    kind='bar',
    x='Product',
    y='Sales',
    legend=False,
    color='skyblue'
)

plt.title("Total Sales by Product")
plt.xlabel("Product")
plt.ylabel("Sales")
plt.show()
```
